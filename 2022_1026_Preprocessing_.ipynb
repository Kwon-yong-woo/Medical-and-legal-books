# 2. extract full sentences only

- "2칸 이상의공백"은 살린다 

```python
original_all = []

for fn_ in fn_list:
    fn_s = globals ()["{}".format(fn_)].splitlines()
    # Rule 2: # strip \t, \xa0 관련 제거하기
    origi_ = [i.strip().replace("\t"," ").replace("\xa0","") for i in fn_s ]
    original_all.extend(origi_)
```

```python
# Rule 1: 2칸 이상의 enter은 space 2칸으로 대체한다.
original_all = re.sub(' {2,}','  ', " ".join(original_all))
```

```python

reg = re.compile(".*?[)][.]|.*?다[.]")
reg = re.compile(" {2,}|.*?다[.]|.*?[\)\.]")
reg = re.compile(".*?[)][.]|.*?다[.]")
```

```python
# Rule 2: sentence split (1. 2칸의 공백 / 2. 다로 끝나는것 / 3. ). 로 끝나는 것
    # 나. 뭐시기 뭐시기 .... 다. 뭐시기 의 다. 같은 경우가 있어서 *이 아니라 +로 쓴다.
reg = re.compile(" {2,}|.+?다[.]|.*?[)][.]")
or_ = reg.findall(original_all)
or_
```

```python
# Rule 3. Sentence split (다로 끝나는 것 중에서 ")."를 포함한것)
or_all = []
for i in or_:
    tar_ = ")."
    if tar_ in i:
        can_ = i.split(tar_)
        can_2 = [i2 +tar_ for i2 in can_[:-1]]
        can_2.append(can_[-1])
        or_all.extend(can_2)
    else:
        or_all.append(i)
```

```python
# Rule 4+ : strip 한번더 하기
or_all = list(map(lambda x: x.strip() if x != "  " else "  ",or_all))
or_all
```

# 3. eliminate short sentences

- 제거 기준 
1. characters
2. space (단어)

```python
# characters 기준으로
ver_ = "cha"
tar_len = 4
sp_ = or_all
def eliminate_short(version,tar_len,sp_):
    tar_sen = []
    remove_sen = []
    for i in sp_:
        if i =="  ":
            tar_sen.append(i)
        else:
            if version =="cha":
                if len(i) >= tar_len:
                    tar_sen.append(i)
                else:
                    remove_sen.append(i)
            elif version =="word":
                if len(i.split(" ")) >= tar_len:
                    tar_sen.append(i)      
                else:
                    remove_sen.append(i)
    return tar_sen, remove_sen
```
