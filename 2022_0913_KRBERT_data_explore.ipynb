```python
def cleansing(text):
    # 숫자의 경우 우선 생략          |(\d+) #숫자
    pattern = """ (\[.*\]) | (\<.*\>) | (【\w+=\w+】) #괄호문자
          |(\d+-\d+-\d+) #전화번호1
          |(\d+\)\d+-\d+) #전화번호2
          |(\w+\@.*\.\w+) #이메일주소
          |(www\.\w+\.[^\s+][\.\w]+) # 링크주소1
          |(http.*[a-zA-Z]) #링크주소2
          |(\w+=\w+) #글자=글자 형태
          |(\w*\s?기자) #기자이름
          |(\d+\.) #숫자목록
          |(\d+\w*\,?\d*원) #금액
          |(\s){2,}
          |[“”:，]
          |[■▦◆▲☎●▶▽△◇→━©ⓒ①②③④⑤⑥⑦®㈜※] #특수문자
          |[(\xa0)] #특수문자2
          |[!#$%&^\'()*+,<=>?@[\\]^_\"{\}~(\/)(\\\\)] #일반적 문장부호
          """

    p = re.compile(pattern, re.VERBOSE)
    tmp = p.sub("", text)
    tmp = tmp.strip()
    return tmp
```
