```python
# 특수문자 등 클린징
def cleansing(text):
    # 숫자의 경우 우선 생략          |(\d+) #숫자
    pattern = """ (\[.*\]) | (\<.*\>) | (【\w+=\w+】) #괄호문자
          |(\d+-\d+-\d+) #전화번호1
          |(\d+\)\d+-\d+) #전화번호2
          |(\w+\@.*\.\w+) #이메일주소
          |(www\.\w+\.[^\s+][\.\w]+) # 링크주소1
          |(http.*[a-zA-Z]) #링크주소2
          |(\w+=\w+) #글자=글자 형태
          |(\w*\s?기자) #기자이름
          |(\d+\.) #숫자목록
          |(\d+\w*\,?\d*원) #금액
          |(\s){2,}
          |[“”:，]
          |[■▦◆▲☎●▶▽△◇→━©ⓒ①②③④⑤⑥⑦®㈜※] #특수문자
          |[(\xa0)] #특수문자2
          |[!#$%&^\'()*+,<=>?@[\\]^_\"{\}~(\/)(\\\\)] #일반적 문장부호
          """

    p = re.compile(pattern, re.VERBOSE)
    tmp = p.sub("", text)
    tmp = tmp.strip()
    return tmp


# 영어 P. R. O. O. F. 와 같은 형식을 PROOF로 변경
def change_eng(text):
    tmp = text
    pattern_eng = "([a-zA-Z]\.\s*)"
    pp = re.compile(pattern_eng)

    find_iter = [f for f in pp.finditer(tmp)]
    tmp_copy = tmp

    for i in range(len(find_iter)):  
        origin = find_iter[i].group()
        changed = find_iter[i].group().replace('. ', '')
        tmp_copy = re.sub(origin,changed,tmp_copy)


    return tmp_copy

def preprocessing(text):
    tmp = text
    tmp = cleansing(tmp)
    tmp = change_eng(tmp)
    tmp = [ t.strip()+'.' for t in tmp.split('.') if t != '' ]
    
#     try:
#         tmp.index('..')
#         drop_id = tmp.index('..')
#         tmp.pop(drop_id)
#     except :
#         pass
    return tmp
    
    
all_article_list = []
all_article_seg = []

for i in tqdm(range(len(body))):
    cleansed_body = preprocessing(body[i])
    all_article_list.extend(cleansed_body)
    all_article_seg.extend([i]*len(cleansed_body))


print(len(all_article_list), len(all_article_seg))
    
# tmp = [ t.strip()+'.' for t in tmp.split('.')]
# |['!"#$%&\'()*+,-/:;<=>?@[\\]^_`{|}~']
# . - / ; : 이런것들은 마침표가 사라짐
```
